```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('D:\Projects\Resume project\dtp2\Data\delivary_time.csv')
df.head()
```

```python
# Univariate Analysis

# 1. Numerical Columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
    
    print(f"Summary statistics for {col}:")
    print(df[col].describe())
    print("\n")

# Observation: Add your observations about the distribution and summary statistics for each numerical column here.

# 2. Categorical Columns
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.show()
    
    print(f"Value counts for {col}:")
    print(df[col].value_counts(normalize=True))
    print("\n")

# Observation: Add your observations about the distribution and value counts for each categorical column here.
```

```python
# Bivariate Analysis

# 1. Numerical vs Numerical
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Observation: Add your observations about the correlations between numerical variables here.

# 2. Categorical vs Numerical
target_variable = 'Time_taken(min)'  # Assuming this is the main variable of interest

for col in categorical_columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=col, y=target_variable, data=df)
    plt.title(f'{col} vs {target_variable}')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Observation: Add your observations about the relationship between each categorical variable and the target variable here.

# 3. Categorical vs Categorical
for col1 in categorical_columns:
    for col2 in categorical_columns:
        if col1 != col2:
            plt.figure(figsize=(12, 6))
            sns.countplot(x=col1, hue=col2, data=df)
            plt.title(f'{col1} vs {col2}')
            plt.xticks(rotation=45, ha='right')
            plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.tight_layout()
            plt.show()

# Observation: Add your observations about the relationships between pairs of categorical variables here.
```

```python
# 1. Time-based Analysis
# Convert 'Time_Orderd' and 'Time_Order_picked' to datetime
df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'])
df['Time_Order_picked'] = pd.to_datetime(df['Time_Order_picked'])

# Calculate time difference between order and pickup
df['Order_to_Pickup_Time'] = (df['Time_Order_picked'] - df['Time_Orderd']).dt.total_seconds() / 60

# Analyze Order_to_Pickup_Time
plt.figure(figsize=(12, 6))
sns.histplot(df['Order_to_Pickup_Time'], kde=True)
plt.title('Distribution of Order to Pickup Time')
plt.xlabel('Time (minutes)')
plt.show()

print("Summary statistics for Order to Pickup Time:")
print(df['Order_to_Pickup_Time'].describe())

# Observation: Add your insights about the order to pickup time distribution
```

```python
# 2. Delivery Person Performance Analysis
avg_time_by_person = df.groupby('Delivery_person_ID')['Time_taken(min)'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_time_by_person.head(20).plot(kind='bar')
plt.title('Average Delivery Time by Delivery Person (Top 20)')
plt.xlabel('Delivery Person ID')
plt.ylabel('Average Time Taken (min)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Observation: Comment on the variation in performance among delivery persons
```

```python
# 3. Impact of Weather Conditions on Delivery Time
plt.figure(figsize=(12, 6))
sns.boxplot(x='Weatherconditions', y='Time_taken(min)', data=df)
plt.title('Impact of Weather Conditions on Delivery Time')
plt.xlabel('Weather Conditions')
plt.ylabel('Time Taken (min)')
plt.xticks(rotation=45, ha='right')
plt.show()

# Observation: Analyze how different weather conditions affect delivery times
```
